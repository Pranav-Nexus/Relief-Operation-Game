```python
import random

# Game variables
resources = 100  # Initial resources
people = 50  # Initial number of survivors
day = 1  # Initial day of the relief operation
rescue_teams = 3  # Initial number of rescue teams
supplies = 50  # Initial amount of supplies
injured_people = 0  # Initial number of injured people
morale = 100  # Initial morale level
damage = 0  # Initial damage level
weather = 'Sunny'  # Initial weather conditions
game_duration = 14  # Total duration of the game in days
tries = 3  # Number of tries allowed
days_to_rescue = 7  # Initial number of days remaining to rescue survivors

# Function to print game state
def print_game_state():
    print("Day:", day)
    print("Resources:", resources)
    print("People:", people)
    print("Rescue Teams:", rescue_teams)
    print("Supplies:", supplies)
    print("Injured People:", injured_people)
    print("Morale:", morale)
    print("Damage:", damage)
    print("Weather:", weather)
    print("Days remaining to rescue survivors:", days_to_rescue)
    print("Tries remaining:", tries)

# Function to handle sending a rescue team
def send_rescue_team():
    global resources, people, morale, rescue_teams, days_to_rescue
    if rescue_teams > 0:
        rescue = random.randint(10, 20)
        resources -= rescue * 5
        people += rescue
        morale += 5
        rescue_teams -= 1
        days_to_rescue -= 1
        print("Sent a rescue team and rescued", rescue, "people")
    else:
        print("No rescue teams available.")

# Function to handle distributing supplies
def distribute_supplies():
    global resources, supplies, people, morale
    if supplies > 0:
        supplies = random.randint(5, 15)
        resources -= supplies * 2
        people += supplies
        morale += 2
        supplies -= 1
        print("Distributed supplies and helped", supplies, "people")
    else:
        print("No supplies available.")

# Function to handle allocating resources for medical treatment
def allocate_resources_for_medical_treatment():
    global resources, morale, injured_people
    if injured_people > 0:
        resources -= injured_people * 3
        morale += 1
        print("Allocated resources for medical treatment.")
        injured_people = 0
    else:
        print("No injured people to treat.")

# Function to handle waiting
def wait():
    global morale, tries
    morale -= 2
    print("Waiting for the situation to stabilize")
    tries -= 1

# Function to update game state
def update_game_state():
    global resources, people, injured_people, morale, damage, weather, day
    resources -= 10  # Resources consumed for daily operations
    people -= random.randint(5, 10)  # People lost due to the tsunami
    injured_people += random.randint(0, 5)  # Randomly generate number of injured people
    morale -= 1  # Morale decreases each day
    damage += random.randint(1, 5)  # Randomly increase damage level
    # Update weather conditions
    weather = random.choice(['Sunny', 'Cloudy', 'Rainy', 'Stormy'])

# Function to check for game over condition
def is_game_over():
    global day, resources, people, tries
    if day > game_duration:  # Check if game duration has exceeded
        print("Game Over: Relief operation duration has ended.")
        return True
    elif resources <= 0:  # Check if resources have depleted
        print("Game Over: Resources have depleted.")
        return True
    elif people <= 0:  # Check if all survivors have perished
        print("Game Over: All survivors have perished.")
        return True
    elif tries <= 0:  # Check if no more tries are remaining
        print("Game Over: No more tries remaining.")
        return True
    else:
        return False
```

```python
while not is_game_over():
    # Display game state
    print_game_state()

    # Get player's action
    action = input("Enter your action (1: Send rescue team, 2: Distribute supplies, 3: Allocate resources for medical treatment, 4: Wait): ")

    # Update game state based on player's action
    if action == '1':
        send_rescue_team()
    elif action == '2':
        distribute_supplies()
    elif action == '3':
        allocate_resources_for_medical_treatment()
    elif action == '4':
        wait()
    else:
        print("Invalid action. Try again.")

    # Update game state
    update_game_state()

    # Increment day counter
    day += 1

# Game over
print("Game Over: Thank you for playing!")
```
